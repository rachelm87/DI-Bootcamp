```python
#Exercise 2: Descriptive Statistics
import scipy.stats as stats

#calculate the mean, median, variance, and standard deviation using SciPy for sample dataset (see below)
data = [12, 15, 13, 12, 18, 20, 22, 21]

mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)

print(f"The mean: {mean_scipy} \n The median: {median_scipy} \n The variance: {variance_scipy} \n The standard deviation: {std_dev_scipy}")
```

```python
#Exercise 3: Understanding Distributions

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

#Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

mu, sigma = 50, 10  # mean and standard deviation
gaussian_dist = stats.norm(mu, sigma)
print(gaussian_dist)
print("PDF at x=50:", gaussian_dist.pdf(50))


x_gaussian = np.linspace(0, 100, 1000) #made it bigger because the top part was not curved at "10" or "100"
pdf_gaussian = gaussian_dist.pdf(x_gaussian)
plt.figure(figsize=(12, 6))
plt.plot(x_gaussian, pdf_gaussian)
plt.title('Gaussian Distribution\n$\mu=50$, $\sigma=10$')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

```python
#Exercise 4: T-Test Application
# Task: Perform a T-test on two sets of randomly generated data.

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)


t_statistic, p_value = stats.ttest_ind(data1, data2)
print("T-statistic:", t_statistic)
print("P-value:", p_value)

#T-Statistic result of -7.316 means that data1  has values significantly lower than data2, whilst the P-value indicates that there's a real difference.
```

```python
#Exercise 5: Linear Regression Analysis - Apply linear regression to a dataset and interpret the results.
import numpy as np
from scipy import stats
#dataset of housing prices (house_prices) and their corresponding sizes (house_sizes)
house_sizes = [50, 70, 80, 100, 120] #(in square meters)
house_prices = [150000, 200000, 210000, 250000, 280000] #(in currency units)

slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)           #linear regression to predict the price of a house given its size.

#What is the slope and intercept of the regression line?
print("Slope:", slope)
print("Intercept:", intercept)
#Predict the price of a house that is 90 square meters. #Y = a + bX
new_house = intercept + slope*90
print(f"${new_house}")
#Interpret the meaning of the slope in the context of housing price
##Answer: As the size increases, so too, does the housing price.
```

```python
#Exercise 6: Understanding ANOVA - Objective: Test understanding of ANOVA and its application.
from scipy import stats
## Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Performing an ANOVA test on the given data.

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value) #What are the F-value and P-value?
print("P-value:", p_value)


#Based on the P-value, do the fertilizers have significantly different effects on plant growth?
##ANSWER: Since P-value < 0.05, we have a low probability of observing differences this large by random chance, and therefore, we can say that at least one of the datasets is statistically different enough.
#Explain what would happen if the P-value were greater than 0.05.
##ANSWER: We could assume the opposite - that there's no meaningful difference and that random chance could be more likely to happen.
```

```python
#Exercise 7: Probability Distributions (Optional) -  Work with a binomial distribution: calculate probabilities for different numbers of successes.
#Calculating the probability of getting exactly 5 heads in 10 coin flips.
from scipy.stats import binom
n, p = 10, .5  # number of trials, probability of each trial
binom_dist = stats.binom(n, p)
print("PMF at 5 successes:", binom_dist.pmf(5))
```

```python
#Exercise 8: Correlation Coefficients (Optional) - Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
import pandas as pd
from scipy.stats import pearsonr
from scipy.stats import spearmanr

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})


# Calculate Pearson Correlation between age and income
corr, _ = pearsonr(data['age'], data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')

# Calculate Spearman Rank Correlation between Age and income
corr, _ = spearmanr(data['age'], data['income'])
print(f'Spearman Rank Correlation: {corr:.2f}')


#As age increases, income increases linearly, and its perfectly aligned (spearman)
```
