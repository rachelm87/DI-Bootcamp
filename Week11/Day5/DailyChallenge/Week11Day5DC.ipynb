```python
#Exploratory Data Analysis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
from IPython.display import display

# Download the dataset
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")


df = pd.read_csv(f"{path}/data.csv")

print(df.shape)
display(df.head())
display(df.info())
display(df.describe())
```

```python
#Clean Up
#checked missing values and dropped
df = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore') #drop last column

df['diagnosis'] = df['diagnosis'].map({'B': 0, 'M': 1}) ##map categorical values to numerical values
print(df['diagnosis'].value_counts()) ##counts of unique rows in the ‘diagnosis’ column

```

```python
#create a countplot
plt.figure(figsize=(15,5))
sns.countplot(y='diagnosis', data=df, palette='magma')

# Enhance the plot with titles and labels
plt.title("# of Patients with Breast Cancer Diagnosis ")
plt.yticks(ticks=[0, 1], labels=["Benign", "Malignant"]) #rename the 0/1
plt.xlabel("Patients (Number)")
plt.ylabel("Diagnosis")
plt.show()
```

#The above countplot shows that 350+ patients were diagnosed with benign breast cancer, while 200+ had a malignant case.

```python
#Data Preprocessing, Building Models and Evaluation :
#Splitting the data into train and test
from sklearn.model_selection import train_test_split

y = df['diagnosis']
X = df.drop(columns=['diagnosis'])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model = LogisticRegression(max_iter=10000, random_state=42) #couldn't converge

# we use .ravel() to transform the y_train data into a 1D array because LogisticRegression() expects the target variable (in this case, y_train) to be a 1D array rather than a column vector or a 2D array.
model.fit(X_train, y_train.values.ravel())

pred = model.predict(X_test) #Implement logistic regression and print the accuracy.

# Calculate accuracy
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.4f}")

#The Logistic Regression Model got 93.86% of all diagnoses correct.
```

```python
from sklearn.metrics import f1_score, recall_score
f1_score = round(f1_score(y_test, pred), 2)
recall_score = round(recall_score(y_test, pred), 2)
print("Sensitivity/Recall for Logistic Regression Model 1 : {recall_score}".format(recall_score = recall_score))
print("F1 Score for Logistic Regression Model 1 : {f1_score}".format(f1_score = f1_score))

#The Logistic Regression Model's recall was .86, which means that it caught 85% of all the malignant (positive) cases. This is very good.

#The F-1 score was 0.91, which means that its a good balance of precision and recall. When it catches it, it's usually right.

    #Implement Random Forests and print the accuracy.
    #Implement Support Vector Machines (SVM) and print the accuracy.
    #Which is the best model ?
```

```python
#Implement K Nearest Neighbours and print the accuracy.

import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification


# Create a KNN classifier model
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model
knn.fit(X_train, y_train)

pred = knn.predict(X_test) #prediction of KNN model

accuracy = accuracy_score(y_test, pred)
print(f'Accuracy: {accuracy}')

#KNN model correctly classified 91 out of every 100 test samples.
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest Classifier with 100 trees
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
#Random Forest correctly predicted 97 out of every 100 patients correctly (either benign or malignant)
```

```python
#Implement Support Vector Machines (SVM) and print the accuracy.

from sklearn import svm
clf = svm.SVC(kernel='linear')

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

#92% accuracy
```

#The Random Forest was by FAR the best of the four options. While all four did very well with 90%+ accuracy, I would absolutely choose the Random Forest.
