```python
#Exercise 1: Exploratory Data Analysis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

uploaded = files.upload() ##upload the file

df = pd.read_csv("dataset_heart.csv")


print(df.shape) ##peek at the data
df.head()
df.info()
df.describe()
```

```python
df.head()
```

```python
class_names = {1: 'No Heart Disease', 2: 'Has Heart Disease'}
print(df['heart disease'].value_counts().rename(index=class_names))
```

```python
from sklearn.model_selection import train_test_split

#remove target column
y = df['heart disease']
X = df.drop('heart disease', axis=1)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print(X_train.shape, X_test.shape)

#There are 270 rows with 216 training examples each with 13 features, 54 test rows with 13 features each.
```

```python
#Exercise 2 - Logistic Regression without Grid Search

from sklearn.linear_model import LogisticRegression

# Increase max_iter to help with convergence
model = LogisticRegression(max_iter=10000)

# we use .ravel() to transform the y_train data into a 1D array because LogisticRegression() expects the target variable (in this case, y_train) to be a 1D array rather than a column vector or a 2D array.

model.fit(X_train, y_train.values.ravel())

pred = model.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, pred))
print("Classification Report:\n", classification_report(y_test, pred))
```

#Evaluation of Model:

## The model achieved an accuracy of 85%, indicating strong overall performance.For example, it  identified 92% of patients with heart disease. However, there were 6 fall positives, and 2 false negatives. Performacne was balanced and there was no bias.

```python
#Exercise 3 - Logistic Regression with Grid Search

# Necessary imports
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np
from sklearn.datasets import make_classification


X, y = make_classification(
    n_samples=1000, n_features=20, n_informative=10, n_classes=2, random_state=42)

# Creating the hyperparameter grid
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space}

# Instantiating logistic regression classifier
logreg = LogisticRegression()

# Instantiating the GridSearchCV object
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)

# Assuming X and y are your feature matrix and target variable
# Fit the GridSearchCV object to the data
logreg_cv.fit(X, y)

# Print the tuned parameters and score
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
## cross-validated accuracy of 85.3% with C = 0.0061 and the lower C slightly improves performance.
```

```python
# Exercise 4: Create an SVM classifier with a linear kernel
from sklearn import svm
clf = svm.SVC(kernel='linear')

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Make predictions
Y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, Y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, Y_pred))
print("Classification Report:\n", classification_report(y_test, Y_pred))

##Evaluation: 85%  accuracy with balanced performance.
```

```python
#Exercise 5: Exercise 5 : SVM with Grid Search
from sklearn import svm
from sklearn.model_selection import GridSearchCV
import numpy as np

# Create the parameter grid

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale']
}

# Instantiate the SVM classifier
svc = svm.SVC()

# Instantiate the GridSearchCV object
svc_cv = GridSearchCV(svc, param_grid, cv=5)

# Fit the GridSearchCV object to your data
svc_cv.fit(X_train, y_train.values.ravel())

# Print the tuned parameters and best score
print("Tuned SVM Parameters: {}".format(svc_cv.best_params_))
print("Best Cross-Validation Score: {:.3f}".format(svc_cv.best_score_))

##The SVM model tuned with GridSearchCV achieved a cross-validated accuracy of 83.8% with the best parameters C = 1, kernel = linear, and gamma = scale, confirming that a linear boundary best fits the dataset.
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# Make sure labels are 0 and 1 for XGBoost
y_train_xgb = y_train.replace({1:0, 2:1}) if set(np.unique(y_train)) == {1,2} else y_train
y_test_xgb  = y_test.replace({1:0, 2:1})  if set(np.unique(y_test))  == {1,2} else y_test

# Manually set hyperparameters
model = XGBClassifier(
    learning_rate=0.1,
    n_estimators=200,
    max_depth=3,
    min_child_weight=2,
    subsample=0.8,
    colsample_bytree=0.8,
    gamma=0.0,
    reg_lambda=1.0,
    reg_alpha=0.0,
    random_state=42,
    eval_metric='logloss'
)

# Train and evaluate
model.fit(X_train, y_train_xgb)
y_pred = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test_xgb, y_pred):.3f}")
print("Confusion Matrix:\n", confusion_matrix(y_test_xgb, y_pred))
print("Classification Report:\n", classification_report(y_test_xgb, y_pred))

#I picked learning_rate=0.1 and n_estimators=200 because a smaller learning rate makes training more stable, while a higher number of trees helps the model learn enough patterns. I used a max_depth of 3 to prevent overfitting since this dataset isnt very large. Setting subsample=0.8 and colsample_bytree=0.8 helps the model generalize better by using only a portion of the data and features for each tree.
```

```python
##My manual XGBoost model achieved solid performance with these parameters. The model trained efficiently, avoided overfitting, etc
```

```python
#Exercise 7: from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

# Make sure labels are 0 and 1 for XGBoost
y_train_xgb = y_train.replace({1:0, 2:1}) if set(np.unique(y_train)) == {1,2} else y_train
y_test_xgb  = y_test.replace({1:0, 2:1})  if set(np.unique(y_test))  == {1,2} else y_test

# Define the base model
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# Define the parameter grid for GridSearchCV
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1]
}

# Set up GridSearchCV
xgb_cv = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)

# Fit the grid search to the training data
xgb_cv.fit(X_train, y_train_xgb)

# Print the best parameters and CV score
print("Best Parameters:", xgb_cv.best_params_)
print(f"Best Cross-Validation Accuracy: {xgb_cv.best_score_:.3f}")

# Evaluate the best model on the test set
best_xgb = xgb_cv.best_estimator_
y_pred = best_xgb.predict(X_test)

print(f"Test Accuracy: {accuracy_score(y_test_xgb, y_pred):.3f}")
print("Confusion Matrix:\n", confusion_matrix(y_test_xgb, y_pred))
print("Classification Report:\n", classification_report(y_test_xgb, y_pred))
```
