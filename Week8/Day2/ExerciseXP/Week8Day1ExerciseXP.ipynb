```python
# Exercise 2: Creating a Line Plot for Temperature Variation
## Create a simple line plot using Matplotlib that represents the temperature variations over a week.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [72, 74, 76, 80, 82, 78, 75]

# Create a new figure
plt.figure()

# Plot x and y
plt.plot(x, y)

# Show the plot
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
plt.title('Temperature variations over a week')
plt.legend()
plt.show()

## Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75])
##and label the x-axis as “Day” and the y-axis as “Temperature (°F)”.

```

```python
## Exercise 3 : Visualizing Monthly Sales with a Bar Chart
### Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.

### Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500])
###and label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.

import matplotlib.pyplot as plt

categories = ['January', 'February', 'March', 'April', 'May']
values = [5000, 5500, 6200, 7000, 7500]

plt.bar(categories, values)
plt.title('Sales Value for Each Month')
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()
```

```python
# Exercise 4 : Data Visualisation
## Using the dataset: sales_data.csv calculate the total quantity of products sold by the company.
from google.colab import files
uploaded = files.upload()

import pandas as pd

df = pd.read_csv("sales_data.csv")
print(df.head(50))  # preview first few rows

##data clean up
mapping = {"Clohting": "Clothing", "Shoeses": "Shoes", "Bgas": "Bags"}
df['category'] = df['category'].replace(mapping).str.strip()

## Identify the category that had the highest revenue and calculate how much revenue it generated.
print(df.columns)
category_revenue = df.groupby('category')['revenue'].sum()
print(category_revenue)
## Determine the average revenue per sale in the dataset.
revenue_sale = df['revenue'].mean()
print("Average Revenue Per Sale:", revenue_sale)

## Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).
## convert category to datetime
print(df['date'].dtype)
df['date'] = pd.to_datetime(df['date'])
print(df.columns) ##confirming!

## extract quarter
df['quarter'] = df['date'].dt.quarter
df['quarter'] = df['quarter'].astype(int)
print(df[['date', 'quarter']].head(100))

## group by total revenue / quarter
total_revenue = df.groupby('quarter')['revenue'].sum()
print("Total Revenue Per Quarter:", total_revenue)

x = total_revenue.index       # quarters
y = total_revenue.values

ax = total_revenue.plot(kind='bar')
ax.set_xlabel("Quarter")
ax.set_ylabel("Revenue")
ax.set_title("Revenue per Quarter")

plt.bar(x, y)
plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.title('Total revenue generated in each quarter of the year')
ax.set_xticklabels([f"Q{int(q)}" for q in total_revenue.index], rotation=0)
plt.show()
```
