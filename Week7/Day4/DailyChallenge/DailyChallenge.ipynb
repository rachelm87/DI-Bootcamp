```python
# Load an image using PIL

from PIL import Image
import matplotlib.pyplot as plt

image_path = 'flowers/flowers/19_010.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)
```

```python
print(type(original_image))
```

```python
## rotate by 90 deg
import numpy as np
from scipy.ndimage import rotate

## convert original image to image array
img_array = np.array(original_image)
print(img_array.shape, img_array.dtype)

## rotate it
rotated_array = rotate(img_array, 90, reshape=False, mode='nearest')

#show it
plt.imshow(rotated_array)
```

```python
## Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
from PIL import ImageOps

horizontally = ImageOps.mirror(original_image)
vertically = ImageOps.flip(original_image)

# Show original
plt.figure()
plt.imshow(original_image)
plt.title("Original")
plt.axis("off")

# Show horizontal flip
plt.figure()
plt.imshow(horizontally)
plt.title("Horizontal Flip")
plt.axis("off")

# Show vertical flip
plt.figure()
plt.imshow(vertically)
plt.title("Vertical Flip")
plt.axis("off")

plt.show()
```

```python
## Zoom in on an image (scale by 1.2x) using .resize(...)

width, height = original_image.size

new_width = int(width * 1.2)
new_height = int(height * 1.2)

bigger_image = original_image.resize((new_width, new_height))

plt.imshow(bigger_image)
```
