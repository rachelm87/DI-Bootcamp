```python
#Exercise 5 - do some math!

import pandas as pd
from statistics import mode
import matplotlib.pyplot as plt

# Calculate mean, median, and mode for SepalLengthCm
mean_value = df_iris["SepalLengthCm"].mean()
median_value = df_iris["SepalLengthCm"].median()
mode_value = df_iris["SepalLengthCm"].mode()[0]  

print("Mean Sepal Length:", mean_value)
print("Median Sepal Length:", median_value)
print("Mode Sepal Length:", mode_value)

df_iris["Species"].value_counts().plot(kind="bar", color="salmon", edgecolor="black")
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

#Results: The mean sepal length is 5.84, the median is 5.8 and the mode is 5.0. Interestingly, it shows 3 different species but I only showed the first 5-6 rows so maybe that's why I didn't realize there were multiple species. 
```
