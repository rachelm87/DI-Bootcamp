```python
#create a new workbook:

from openpyxl import Workbook
from openpyxl.styles import Font, Alignment, PatternFill
import pandas as pd
from openpyxl.utils.dataframe import dataframe_to_rows

#create new workbook and sheet titled 'grades'
wb = Workbook()
ws = wb.active
ws.title = "Grades"

#populate the sheet with the following data structure:
##Column A: Student Names (e.g., Joe, Bill, Tim, Sally, Jane)
##Columns B, C, D, E: Subject Grades (math, science, english, gym)

# Insert headers into "DataSheet"
ws['A1'] = 'Student Name'
ws['B1'] = 'Math Grade'
ws['C1'] = 'Science Grade'
ws['D1'] = 'English Grade'
ws['E1'] = 'Gym Grade'

#format the headers
center_align = Alignment(horizontal='center', vertical='center')
fill = PatternFill(start_color="FF224677", end_color="FF224677", fill_type="solid")
bold_font = Font(bold=True, color="FFFFFFFF")
headers = ["A1","B1","C1","D1","E1"]

for h in headers:
  ws[h].alignment = center_align
  ws[h].fill = fill
  ws[h].font = bold_font

# Create a DataFrame
data = {
    'Student Name': ['Joe', 'Bill', 'Tim', 'Sally', 'Jane'],
    'Math Grade': [65, 55, 100, 30, 100],
    'Science Grade': [78, 72, 45, 25, 100],
    'English Grade':[98, 87, 75, 45, 100],
    'Gym Grade':[89, 95, 92, 100, 60]
    }
df = pd.DataFrame(data)
#add to the new sheet
for row in dataframe_to_rows(df, index=False, header=False):
  ws.append(row)
```

```python
from openpyxl.utils import get_column_letter

first_row = 2
last_row = 6

avg_row = last_row + 1

for col in range(2,6):
  col_letter = get_column_letter(col)
  ws[f"{col_letter}{avg_row}"] = f"=AVERAGE({col_letter}{first_row}:{col_letter}{last_row})"


ws[f"A{avg_row}"] = "Average Grade"
```

```python
wb.save("Grades.xlsx")

df_check=pd.read_excel("Grades.xlsx")
df_check
```
