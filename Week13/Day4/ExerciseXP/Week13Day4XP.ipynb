```python
#Exercise 5 : Analyzing US-CERT Security Alerts

##Write a Python program to get the number of security alerts issued by US-CERT in the current year.

import requests
from bs4 import BeautifulSoup
import datetime

this_year = datetime.datetime.now().year

url = "https://www.cisa.gov/news-events/cybersecurity-advisories?f%5B0%5D=advisory_type%3A93"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.raise_for_status()

soup = BeautifulSoup(response.text, "html.parser")

count = 0

alert = soup.find_all("article", class_="is-promoted c-teaser c-teaser--horizontal")

for a in alert:
  timeframe = a.find("time")
  if timeframe:
    dateframe = timeframe.get("datetime")

    if str(this_year) in dateframe:
      count += 1

print(f"Total Alerts in 2025: {count}.")
```

```python
#Exercise 6: Scraping Movie Details
##Write a Python program to get movie name, year and a brief summary of the top 10 random movies from this IMBD website.

import requests
from bs4 import BeautifulSoup
import random

#get link
url = "https://www.imdb.com/list/ls091294718/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.raise_for_status()

soup = BeautifulSoup(response.text, "html.parser")


movies = soup.find_all("h3", class_="ipc-title__text ipc-title__text--reduced")

random_movies = random.sample(movies, 10)

for m in random_movies:
    print(m.get_text(strip=True))


```
